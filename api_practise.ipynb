# 2) API calls naar BAG API

Use a list of (Dutch) postal codes (4 digits, 2 chars) to call the BAG API and get the corresponding  'wijk' and 'buurt' (neighborhood and district)

See a complete list of API's + swagger docs of the City of Amsterdam here
https://api.data.amsterdam.nl/api/

```python
pc_list = ['1069KJ', '1058NH', '1057PV'] # f.i.
num_url = """https://api.data.amsterdam.nl/bag/nummeraanduiding/"""

# totaal resultaten..
results = []

for pc_code in pc_list:
    
    params = {'postcode': pc_code, 'detailed' :1}
    response = requests.get(num_url, params)
    data = response.json()
    # LETOP meerdere nummeraanduiding per pc dus ook meerdere buurten
    # mogelijk. grenzen lopen niet mooi..
    result = {'postcode': pc_code}
    if not data.get('results'):
        results.append(result)
        # nothing found.
        continue

    result['buurt'] = data['results'][0]['buurt']['_display']
    result['sd'] = data['results'][0]['gebiedsgerichtwerken']['naam']
    result['sd_code'] = data['results'][0]['gebiedsgerichtwerken']['code']
    result['buurt_combi'] = data['results'][0]['buurtcombinatie']['naam']
    result['voll_code']= data['results'][0]['buurtcombinatie']['vollcode']

    # voeg postcode  resultaat toe aan totaal resultaten lijst.
    results.append(result)

pprint.pprint(results)
```

```python
def query_distances(stations=stations):
    """Query the Google API for bicycling distances"""
    latlon_list = ['{0},{1}'.format(lat, long)
                   for (lat, long) in zip(stations.lat, stations.long)]

    def create_url(i):
            URL = ('https://maps.googleapis.com/maps/api/distancematrix/json?'
                   'origins={origins}&destinations={destinations}&mode=bicycling'
                  '&key=' + str(api_key))
            return URL.format(origins=latlon_list[i],
                              destinations='|'.join(latlon_list[i + 1:]))
    
    for i in range(len(latlon_list) - 1):
        url = create_url(i)
        filename = "distances_{0}.json".format(stations.terminal.iloc[i])
        print (i, filename)
        !curl "{url}" -o {filename}
        sleep(10)
```
